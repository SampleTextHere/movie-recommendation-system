```python
import pandas as pd 

movies = pd.read_csv("data/movies.csv")
ratings = pd.read_csv("data/ratings.csv")

print("Movies:")
display(movies.head(10))

print("Ratings:")
display(ratings.head(10))
```

```python
user_movie_matrix = ratings.pivot_table(
    index='userId',
    columns='movieId',
    values='rating'
)

display(user_movie_matrix.head(10))
```

```python
user_movie_filled = user_movie_matrix.fillna(0)

display(user_movie_filled.head(10))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity = cosine_similarity(user_movie_filled)

user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_movie_matrix.index,
    columns=user_movie_matrix.index
)

display(user_similarity_df.head(5))
```

```python
def get_similar_users(user_id, n=10):

    if user_id not in user_similarity_df.index:
        return "Invalid user ID"

    similar_users = user_similarity_df[user_id].sort_values(ascending=False)

    similar_users = similar_users.iloc[1:n+1]

    return similar_users
```

```python
get_similar_users(1, 5)
```

```python
movie_rating_counts = ratings.groupby('movieId').size()
```

```python
def recommend_movies(user_id, n_movies=10, min_ratings=20):

    if user_id not in user_movie_matrix.index:
        return "Invalid user ID"
    
    similar_users = get_similar_users(user_id)

    similar_users_ratings = user_movie_matrix.loc[similar_users.index]

    mean_ratings = similar_users_ratings.mean(axis=0, skipna=True)

    watched_movies = user_movie_matrix.loc[user_id].dropna().index

    recommendations = mean_ratings.drop(watched_movies)

    recommendations = recommendations[
        movie_rating_counts[recommendations.index] >= min_ratings
    ]

    recommendations = recommendations.sort_values(ascending=False).head(n_movies)

    recommended_movies = movies[movies['movieId'].isin(recommendations.index)][['movieId', 'title']].copy()

    recommended_movies['predicted_rating'] = recommended_movies['movieId'].map(recommendations).round(4)

    recommended_movies = recommended_movies.sort_values(by='predicted_rating', ascending=False).reset_index(drop=True)

    return recommended_movies

```

```python
recommend_movies(1, n_movies=10, min_ratings=20)
```

```python
def weighted_recommend_movies(user_id, n_movies=10, min_ratings=20):

     if user_id not in user_movie_matrix.index:
        return "Invalid user ID"
     
     similar_users = get_similar_users(user_id)

     similarity_scores = similar_users.values

     similar_users_ratings = user_movie_matrix.loc[similar_users.index]

     weighted_ratings  = similar_users_ratings.mul(similarity_scores, axis=0)

     weighted_sum = weighted_ratings.sum(axis=0)

     similarity_sum = (
         similar_users_ratings.notna()
         .mul(similarity_scores, axis=0)
         .sum(axis=0)
     )

     weighted_average = weighted_sum / similarity_sum

     watched_movies = user_movie_matrix.loc[user_id].dropna().index
     
     recommendations = weighted_average.drop(watched_movies)

     recommendations = recommendations[
        movie_rating_counts[recommendations.index] >= min_ratings
    ]
     
     recommendations = recommendations.sort_values(ascending=False).head(n_movies)

     recommended_movies = movies[movies['movieId'].isin(recommendations.index)][['movieId', 'title']].copy()

     recommended_movies['predicted_rating'] = recommended_movies['movieId'].map(recommendations).round(4)
     
     recommended_movies = recommended_movies.sort_values(by='predicted_rating', ascending=False).reset_index(drop=True)

     return recommended_movies

     
```

```python
user_mean_ratings = user_movie_matrix.mean(axis=1)

display(user_mean_ratings.head(10))
```

```python
normalized_matrix = user_movie_matrix.sub(user_mean_ratings, axis=0)

display(normalized_matrix.head(10))
```

```python
normalized_filled = normalized_matrix.fillna(0)

display(normalized_filled.head(10))
```

```python
normalized_similarity = cosine_similarity(normalized_filled)

normalized_similarity_df = pd.DataFrame(
    normalized_similarity,
    index=user_movie_matrix.index,
    columns=user_movie_matrix.index
)

display(normalized_similarity_df.head(10))
```

```python
def get_normalized_similar_users(user_id, n=10):

    if user_id not in normalized_similarity_df.index:
        return "Invalid user ID"

    similar_users = normalized_similarity_df[user_id].sort_values(ascending=False)

    similar_users = similar_users.iloc[1:n+1]

    return similar_users
```

```python
def normalized_weighted_recommend_movies(user_id, n_movies=10, min_ratings=20):

    if user_id not in user_movie_matrix.index:
        return "Invalid user ID"
    
    similar_users = get_normalized_similar_users(user_id, n=50)

    similarity_scores = similar_users.values

    similar_users_ratings = normalized_matrix.loc[similar_users.index]

    weighted_ratings = similar_users_ratings.mul(similarity_scores, axis=0)

    weighted_sum = weighted_ratings.sum(axis=0)

    similarity_sum = (
        similar_users_ratings.notna()
        .mul(similarity_scores, axis=0)
        .sum(axis=0)
        )

    predicted_scores = weighted_sum / similarity_sum

    predicted_scores = predicted_scores + user_mean_ratings[user_id]

    predicted_scores = (predicted_scores / predicted_scores.max()) * 5

    predicted_scores = predicted_scores.clip(0.5, 5)

    watched_movies = user_movie_matrix.loc[user_id].dropna().index

    recommendations = predicted_scores.drop(watched_movies)

    recommendations = recommendations[
        movie_rating_counts[recommendations.index] >= min_ratings
        ]

    recommendations = recommendations.sort_values(ascending=False).head(n_movies)

    recommended_movies = movies[movies['movieId'].isin(recommendations.index)][['movieId', 'title']].copy()

    recommended_movies['predicted_rating'] = recommended_movies['movieId'].map(recommendations).round(4)

    recommended_movies = recommended_movies.sort_values(by='predicted_rating', ascending=False).reset_index(drop=True)

    return recommended_movies
```
